```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train = pd.read_csv('data.csv')
```

```python
train
```

# EDA And Feature Engineering

```python
train.info()
```

```python
train['date'].unique()
```

NOTE: Since all belong to the same year, just taking the month and date of reading will be fine

```python
train["date"] = train["date"].astype('string')
```

```python
train["month"] = train["date"].apply(lambda x: (x.split()[0]).split("-")[1])
```

```python
train["date"] = train["date"].apply(lambda x: (x.split()[0]).split("-")[2])
```

```python
train
```

NOTE: yr_renovated is same as yr_built in case of 0

```python
train.loc[train['yr_renovated'] == 0, 'yr_renovated'] = train['yr_built']
```

```python
train
```

Removing the number of the street

```python
train["street"] = train["street"].apply(lambda x: " ".join(x.split()[1:]))
```

```python
train
```

```python
train['statezip'].astype(str).apply(lambda x :x.split()[0]).unique()
```

```python
train['statezip'] = train['statezip'].astype(str).apply(lambda x:x.split()[1])
```

```python
train
```

```python
train['country'].unique()
```

```python
train.drop(columns = ['country'], inplace=True)
```

```python
train
```

```python
train['street'].nunique()
```

```python
#Count Encoding street, since the number of houses on a street can you provide a lot of information
train['street'] = train['street'].map(train['street'].value_counts().to_dict())
```

```python
train['city'].nunique()
```

```python
train['city'] = train['city'].map(train['city'].value_counts().to_dict())
```

```python
train.head()
```

```python
train = train.astype(float)
```

```python
train.isna().sum().sum()
```

```python
train.columns
```

I know street is important, but thinking of more strategies will require time, which I don't have right now. So just gonna drop it for now

```python
X = train[['date', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated', 'street', 'city',
       'statezip', 'month']]
y = train['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7286)
```

# Models

```python
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.linear_model import (Ridge, ElasticNet, SGDRegressor, BayesianRidge,
                                LinearRegression, Lasso, LassoLars, PassiveAggressiveRegressor,
                                HuberRegressor, TheilSenRegressor, RANSACRegressor)
from sklearn.ensemble import (RandomForestRegressor, GradientBoostingRegressor,
                            AdaBoostRegressor, BaggingRegressor, ExtraTreesRegressor)
from sklearn.tree import DecisionTreeRegressor, ExtraTreeRegressor
from sklearn.neighbors import KNeighborsRegressor, RadiusNeighborsRegressor
from sklearn.neural_network import MLPRegressor
```

```python
models_list = {
    'SVR': SVR(),
    'XGBRegressor': XGBRegressor(),
    'SGDRegressor': SGDRegressor(),
    'BayesianRidge': BayesianRidge(),
    'LinearRegression': LinearRegression(),
    'RandomForestRegressor': RandomForestRegressor(),
    'AdaBoostRegressor': AdaBoostRegressor(),
    'BaggingRegressor': BaggingRegressor(),
    'ExtraTreesRegressor': ExtraTreesRegressor(),
    'KNeighborsRegressor': KNeighborsRegressor(),
    'MLPRegressor': MLPRegressor()
}
```

```python
model_rmse = []
model_names = []
model_r2 = []
for name, model in models_list.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)

    model_rmse.append(rmse)
    model_names.append(name)
    model_r2.append(r2)
```

```python
results = {
    'Name': model_names,
    'RMSE': model_rmse,
    'R2': model_r2
}
```

```python
resultsdf = pd.DataFrame(results).sort_values(by='RMSE', ascending=False)
```

```python
resultsdf
```

Best model is bagging model of ExtraTreesRegressor

## ExtraTreesRegressor

```python
extra_trees = ExtraTreesRegressor(random_state=10)

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt'],
}

grid_search = GridSearchCV(estimator=extra_trees, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)

# Use the best model for predictions
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print("Root Mean Squared Error (RMSE):", rmse)
print("R2 Score:", r2)
```
